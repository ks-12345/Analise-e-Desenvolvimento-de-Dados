```python
# Para rodar a aplicao
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurao visual
sns.set_theme(style="darkgrid")

# Importando o arquivo
df = pd.read_csv('supermercado.csv')

# Mostrar as primeiras 5 linhas
display(df.head())

# Verificar tipos de dados
df.info()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Region', y='Quantity', palette='Set2')
plt.title('Quantidade por Regio')
plt.xlabel('Regio')
plt.ylabel('Quantidade')
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=30, kde=True, color='skyblue')
plt.title('Distribuio das Vendas')
plt.xlabel('Vendas (Sales)')
plt.ylabel('Contagem')
plt.show()


valores = df['Segment'].value_counts()
labels = valores.index

plt.figure(figsize=(8, 8))
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribuio por Segmento')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Quantity', estimator='sum', palette='pastel')
plt.title('Total de Quantidade por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Total')
plt.show()


plt.figure(figsize=(8, 5))
sns.heatmap(df[['Sales','Quantity','Discount','Profit']].corr(), annot=True, cmap='Blues')
plt.title('Correlao entre Variveis Numricas')
plt.show()
```
